```python
#read file dowloaded from kaggle.com
df=pd.read_csv(r"C:\Users\user\Desktop\PortfolioProject2023\London Bike Rides Analysis\london_merged.csv")
df
```

```python
#count of unique values in weather_code
df.groupby('weather_code').count()
```

```python
#count of unique values in season
df.groupby('season').count()
```

```python
#Renaming column
new_column_dict = {
    'timestamp':'time',
    'cnt':'count', 
    't1':'real_temp_C',
    't2':'apparent_temp_C',
    'hum':'humidity_percent',
    'wind_speed':'wind_speed_kph',
    'weather_code':'weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}

df.rename(columns=new_column_dict, inplace=True)
```

```python
#changing humidity values to percentage
df['humidity_percent']=df['humidity_percent'].apply(lambda x: x / 100)
```

```python
#Changing values in weather and season column into string
#Info on what those int represent are in kaggle.com -> https://www.kaggle.com/datasets/hmavrodiev/london-bike-sharing-dataset

#Creating dictionaries
weather_dict = {
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0':'Broken clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall' 
}

season_dict = {
    '0.0':'spring',
    '1.0':'summer',
    '2.0':'autumn',
    '3.0':'winter'
}
```

```python
#Changing data type from int to string
df['weather']=df['weather'].astype('str')
df['season']=df['season'].astype('str')
```

```python
#Changing values
df['weather']=df['weather'].map(weather_dict)
df['season']=df['season'].map(season_dict)
```
